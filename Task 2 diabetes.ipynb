```python
#Distribution of Age
diabetes.Age.plot(color="red",kind="hist")
plt.title("Distribution of Age" , {'fontsize':14,'color':'darkblue'})
plt.show()
```

```python
#Scatter plot for Age and Glucose
plt.scatter(data= diabetes ,x='Age' , y= 'Glucose' , c= 'Outcome',lw = 0.5)
plt.title("Age vs Glucose" , {'fontsize':14,'color':'r'})
```

```python
#pie chart for outcome
plt.pie(diabetes['Outcome'].value_counts(),labels=[0,1] , shadow= True ,explode= [.1,.1] , autopct="%1.1f%%"  )
plt.legend(loc="upper right" )
plt.title("Outcome" , {'fontsize':14,'color':'darkgray'})
```

```python
#Body Mass Index bocplot
diabetes['BMI'].plot(kind="box", color="red", vert=False)
plt.show()
```

```python
sns.heatmap(diabetes.corr(),annot=True,cmap="inferno",lw=0.5 ,ls="solid")
```

```python

plt.scatter(data= diabetes ,x='Age' , y= 'Pregnancies' , c= 'Outcome',lw = 0.5)
plt.title("Age vs Pregnancies" , {'fontsize':14,'color':'darkblue'})
plt.xlabel("Age")
plt.ylabel("Pregnancies")
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.barplot(data= diabetes ,x='Insulin' , y= 'SkinThickness' )
plt.title("Insulin and the correlation with SkinThickness" , {'fontsize':14,'color':'darkblue'})
plt.show()
```

```python
sns.histplot(data=diabetes['Glucose'],color='r',kde=True)
plt.title('Glucose in Dataset', fontsize=15, color='darkblue')
```

```python
#bmi vs age hued by outcome
plt.figure(figsize=(10, 6))
sns.scatterplot(x='BMI', y='Age', hue='Outcome', data=diabetes, palette='Set1', s=90)
plt.title('BMI vs Age hued by Outcome')
plt.show()
```

```python
diabetes.hist(bins=20, figsize=(10, 10), color='darkblue')
plt.suptitle('Feature Distribution')
plt.show()
```

```python
#scatter plot of Pregnancies and Outcome
plt.subplot(2, 1, 1)
plt.plot(diabetes['Pregnancies'], diabetes['Outcome'],'rs')


plt.subplot(2, 1, 2)
plt.plot(diabetes['Age'], diabetes['Outcome'],'yo')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data=diabetes[diabetes['Outcome'] == 0]['BMI'], label="No Diabetes", shade=True, color="red")
sns.kdeplot(data=diabetes[diabetes['Outcome'] == 1]['BMI'], label="Diabetes", shade=True, color="yellow")
plt.title("Density Plot of BMI by Diabetes Outcome")
plt.xlabel("BMI")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
numerical_features = ['Glucose', 'BloodPressure', 'BMI', 'Age']
plt.figure(figsize=(8, 10))
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data=diabetes, x=feature, kde=True, hue='Outcome', multiple='stack')
    plt.title(f'Distribution of {feature} by Outcome')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.countplot(x='Pregnancies', hue='Outcome', data=diabetes)
plt.title('Number of Pregnancies by Diabetes Outcome')
plt.xlabel('Number of Pregnancies')
plt.ylabel('Count')
plt.legend(title='Outcome', labels=['No Diabetes', 'Diabetes'])
plt.show()
```

```python
sns.pairplot(diabetes, vars=['Glucose', 'BMI', 'Age', 'BloodPressure'], hue='Outcome', diag_kind='kde')
plt.suptitle('Pairplot of Key Features by Diabetes Outcome', y=1.02)
plt.show()
```
